#### Bloch representation

The idea behind this representation of a Bubit is that a Qubit has a state 
-  |0⟩ if it’s amplitude lies at the top of the sphere
-  |1⟩ if it lies at the bottom

Quantum mechanics is mathematically formulated in Hilbert space or projective Hilbert space. 

The pure states of a quantum system correspond to the one-dimensional subspaces of the corresponding Hilbert space (and the "points" of the projective Hilbert space). 

For a two-dimensional Hilbert space, the space of all such states is the complex projective line.

This is the Bloch sphere, which can be mapped to the Riemann sphere.
(source : https://en.wikipedia.org/wiki/Bloch_sphere)

```python

# qubit modelisation and representation
bloch_sphere = qutip.Bloch()
bloch_sphere.show()
```

#### Amplitudes and vectors

A Qubit $|\psi\rangle$ is often represented by the following formula:  

$\vert\psi\rangle = a\vert 0\rangle + b\vert 1\rangle$ 

or

$\vert\psi\rangle = a\vert\uparrow\rangle + b\vert\downarrow\rangle$ 


Here $(a,b)$  are Qubit-coordinates on the Bloch sphere according to the states $\vert 0\rangle$ and $\vert 1\rangle$

Bloch sphere is a unit 2-sphere of $\R^3$ then we have $\langle\psi\vert\langle\psi = \lVert \psi \rVert ^2 = 1$

Take a vector $\vec{a}$ of the Bloch sphere. For example  $\vec{a}=\left(\frac{1}{\sqrt(3)},\frac{1}{\sqrt(3)},\frac{1}{\sqrt(3)}\right)$
  
Here a way to represent it on Bloch sphere (source : https://qutip.org/docs/latest/guide/guide-bloch.html?highlight=basis)

```python
# adding the vector a on the Bloch sphere
vector_a = [1./np.sqrt(3), 1./np.sqrt(3), 1./np.sqrt(3)]
bloch_sphere.add_points(vector_a)

# adding a vector y 
vec = [0, 1, 0]
bloch_sphere.add_vectors(vec)

# adding the up-state 
up = qutip.basis(2, 0)
bloch_sphere.add_states(up)

bloch_sphere.render()
bloch_sphere.show()
```

What is important here : a Qubit is a superposition of states. 

$=>$ The Qubit holds both states even in our example it seems closer to the $\vert 0\rangle$

$=>$ It is NOT a binary state like $0$ or $1$

The main consequence of this states superposition is the following fact :

$=> \space x$ Qubits can represent ALL values between 0 & $2^x$ at a time 

While

$=> \space x$ bits can represent ONE value between 0 & $2^x$ at a time
